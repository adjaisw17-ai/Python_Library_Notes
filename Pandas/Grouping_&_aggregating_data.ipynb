```python
# calculating how many people in each country know python
Total_respondents = df['Country'].value_counts()
Total_respondents
```

```python
pwkPython= country_df['LanguageHaveWorkedWith'].apply(lambda x : x.str.contains('Python').sum())
pwkPython.head()
```

```python
# creating a new df 
calculate_pcwkpython = pd.concat([pwkPython,Total_respondents],axis='columns')
calculate_pcwkpython
```

```python
calculate_pcwkpython.rename(columns = {'LanguageHaveWorkedWith':'Know Python' , 'count' : 'total respondents'},inplace=True)
```

```python
calculate_pcwkpython['percantage_know_python'] = calculate_pcwkpython['Know Python']/calculate_pcwkpython['total respondents']*100
calculate_pcwkpython
```

```python
calculate_pcwkpython.sort_values('percantage_know_python',ascending =False)
```
