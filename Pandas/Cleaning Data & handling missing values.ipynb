```python
import pandas as pd
import numpy as np
```

```python
people ={
    'first' : ['Corey','Anand','Aditya','Ashish','Prateek',np.nan,None,'NA'],
    'last' : ['Schafar','Jaiswal','Jaiswal','Jaiswal','Rastogi',np.nan,np.nan,None],
    'email' : ['corey@gmail.com','anand@gmail.com','aditya@gmail.com',None,'prateek@gmail.com',None,np.nan ,'missing@gmail.com'],
     'age' : ['33','18','18','19','20',None,None,None]
}
```

```python
df =pd.DataFrame(people)
df.replace('NA',np.nan,inplace=True)
df.replace('Missing',np.nan,inplace=True)
```

```python
df
```

```python
df.dropna()  #drops all rows having any  missing column
```

```python
df.dropna(axis='index',how ='any')  # this is same as above by default function of dropna function
```

```python
df.dropna(axis='index',how='all')   # this will remove only those rows having all values missing
```

```python
df.dropna(axis='columns',how ='all')
```

```python
df.dropna(axis='columns',how ='any')
```

```python
df.dropna(axis='index',how ='any',subset =['email'])  # this gives rows which have  email in them
```

```python
df.dropna(axis='index',how ='any',subset =['email','last'])  # remove those which do not have either the last or the email
```

```python
df.dropna(axis='index',how ='all',subset =['email','last'])  # remove those not having both last and email
```

```python
df.isna()
```

```python
df.fillna(0)
```
