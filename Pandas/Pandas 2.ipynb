```python
people ={ 
      'first': ["Corey","Anand","Ashish"],
    'last'  : ["Schafar" ,"Jaiswal","Jaiswal"],
    'email' : ["corey@gmail.com" ,"anand@gmail.com","ashish@gmail.com"]
}
```

```python
import pandas as pd
df =pd.DataFrame(people)
df
```

```python
# adding new column in a dataframe
```

```python
full_name =df['first']+" "+df['last']
df['fullname']=full_name
df
```

```python
# deleting columns in a dataframe
```

```python
df.drop(columns =['first','last'])
```

```python
# if we want to make this change permanent use inplace =True
```

```python
# spliting parts of a string
```

```python
df['fullname'].str.split(' ')
```

```python
# adding these splited values as two new columns in a dataframe
```

```python
df['fullname'].str.split(' ',expand =True) # this makes columns of splitted values
```

```python
df[['first','last']]=df['fullname'].str.split(' ',expand =True)
```

```python
df
```

```python
# adding rows   -  use concat function - it can concat rows as well as two dataframes
```

```python
new_row =pd.DataFrame([{'first':'Bruce'}])
df=pd.concat([df,new_row],ignore_index=True)
```

```python
df
```

```python
#Joining two dataframe end to end to make a new DataFrame
```

```python
people ={ 
      'first': ["Steve","Tony"],
    'last'  : ["Rogers" ,"Stark"],
    'email' : ["cap@gmail.com" ,"ironman@gmail.com"]
}
df1 =pd.DataFrame(people)
```

```python
#vertical concatenation - i.e, joining by rows
df2 =pd.concat([df,df1],ignore_index = True)  
# here ignore index helps to reorder indices according 
```

```python
df2
```

```python
#horizontal concatenation - joining by columns
df3=pd.concat([df,df1],axis='columns')  # we can also use axis =1 to join by column ,axis=0 → stack rows (default)axis=1 → add columns side-by-side.
```

```python
df3
```

```python
# removing rows from dataframe
```

```python
df3.drop(index=3)
```
